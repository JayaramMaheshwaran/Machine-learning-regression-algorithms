```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Function to load the dataset
def importdata():
    # Replace with your dataset path
    data = pd.read_csv(r"C:\Users\Admin\OneDrive\Desktop\task\class notes\algorithm\1..linear\bostonhousing.csv")
    return data

# Function to split the dataset
def splitdataset(data):
    X=boston.loc[:,['CRIM','ZN','INDUS']] 
    y=boston.loc[:,'MEDV']
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.15)
    return X_train, X_test, y_train, y_test

# Function to train and evaluate Linear Regression
def linear_regression(X_train, X_test, y_train, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    return mse, r2

# Function to train and evaluate K-Nearest Neighbors
def knn(X_train, X_test, y_train, y_test):
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    model = KNeighborsRegressor(n_neighbors=5)
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    return mse, r2

# Function to train and evaluate Decision Tree
def decision_tree(X_train, X_test, y_train, y_test):
    model = DecisionTreeRegressor(random_state=10)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    return mse, r2

# Function to train and evaluate Random Forest
def random_forest(X_train, X_test, y_train, y_test):
    model = RandomForestRegressor(n_estimators=100, random_state=10)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    return mse, r2

# Function to train and evaluate XGBoost
def xgboost(X_train, X_test, y_train, y_test):
    model = XGBRegressor(objective='reg:squarederror', random_state=10)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    return mse, r2

# Function to train and evaluate Support Vector Machine
def svm(X_train, X_test, y_train, y_test):
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    model = SVR()
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    return mse, r2

# Main function to run all models and plot the results
def driver():
    data = importdata()
    X_train, X_test, y_train, y_test = splitdataset(data)
    
    # Linear Regression
    lr_mse, lr_r2 = linear_regression(X_train, X_test, y_train, y_test)
    
    # K-Nearest Neighbors
    knn_mse, knn_r2 = knn(X_train, X_test, y_train, y_test)
    
    # Decision Tree
    dt_mse, dt_r2 = decision_tree(X_train, X_test, y_train, y_test)
    
    # Random Forest
    rf_mse, rf_r2 = random_forest(X_train, X_test, y_train, y_test)
    
    # XGBoost
    xgb_mse, xgb_r2 = xgboost(X_train, X_test, y_train, y_test)
    
    # Support Vector Machine
    svm_mse, svm_r2 = svm(X_train, X_test, y_train, y_test)
    
    # Collecting R^2 scores for plotting
    algorithms = ['Linear Regression', 'K-Nearest Neighbors', 'Decision Tree', 'Random Forest', 'XGBoost', 'Support Vector Machine']
    r2_scores = [lr_r2, knn_r2, dt_r2, rf_r2, xgb_r2, svm_r2]
    
    # Printing MSE and R^2 scores
    def print_results(model_name, mse, r2):
        print(f"{model_name}:\n")
        print("Mean Squared Error:")
        print(mse)
        print("\nR^2 Score:")
        print(r2)
        print("\n--------------------------------------------------------\n")
    
    print_results("Linear Regression", lr_mse, lr_r2)
    print_results("K-Nearest Neighbors", knn_mse, knn_r2)
    print_results("Decision Tree", dt_mse, dt_r2)
    print_results("Random Forest", rf_mse, rf_r2)
    print_results("XGBoost", xgb_mse, xgb_r2)
    print_results("Support Vector Machine", svm_mse, svm_r2)
    
    # Plotting the bar chart for R^2 scores
    plt.figure(figsize=(15, 6))
    bars = plt.bar(algorithms, r2_scores, color=['blue', 'green', 'red', 'purple', 'orange', 'cyan'])
    plt.xlabel('Algorithms')
    plt.ylabel('R^2 Score')
    plt.title('Comparison of Algorithm R^2 Scores')
    plt.ylim(0, 1)
    for bar, r2 in zip(bars, r2_scores):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 0.05, f'{r2:.2f}', ha='center', color='white', fontsize=10)
    plt.show()

if __name__ == "__main__":
    driver()
```
